```python
import pandas as pd
import os

raw_path = "data/raw/Fashion_Dataset_v2.csv"  # adjust filename if different

df = pd.read_csv(raw_path)
df.head()
df.info()
df.describe(include="all")
```

```python
# Keep only useful columns (adjust based on actual dataset)
cols = [
    "p_id",
    "name",
    "products",
    "price",
    "colour",
    "brand",
    "ratingCount",
    "avg_rating",
    "description",
    "p_attributes",
]
df = df[cols].copy()

df = df.rename(
    columns={
        "p_id": "product_id",
        "name": "productDisplayName",
        "colour": "baseColour",
    }
)

# Handle missing values
# 👉 Handle missing values defensively (even though this file is quite clean)
text_cols = [
    "productDisplayName",
    "products",
    "baseColour",
    "brand",
    "description",
    "p_attributes",
]
for col in text_cols:
    df[col] = df[col].fillna("")

# Numeric columns: ratingCount, avg_rating, price
df["ratingCount"] = df["ratingCount"].fillna(0).astype(int)
df["avg_rating"] = df["avg_rating"].fillna(0).astype(float)

# Ensure price is numeric and drop rows without price (critical field)
df["price"] = pd.to_numeric(df["price"], errors="coerce")
df = df.dropna(subset=["price"])
df["price"] = df["price"].astype(float)

```

```python
print(df.head())
```

```python
print(df.dtypes)
```

```python
#Create raw_text_for_embedding
def build_text(row):
    parts = [
        row["productDisplayName"],
        f"Gender: {row['gender']}",
        f"Category: {row['masterCategory']} / {row['subCategory']} / {row['articleType']}",
        f"Color: {row['baseColour']}",
        f"Usage: {row['usage']}",
        f"Season: {row['season']} {row['year']}",
        f"Price: {row['price']}"
    ]
    return ". ".join([str(p) for p in parts if pd.notna(p)])

df["raw_text"] = df.apply(build_text, axis=1)

df[["productDisplayName", "raw_text"]].head()
```

```python
#Create raw_text_for_embedding
def build_text(row):
    """
    Build a rich, human-like description string for each product,
    combining all useful fields. This is what we’ll embed.
    """
    parts = []

    # Name + brand
    if row["productDisplayName"]:
        parts.append(str(row["productDisplayName"]))
    if row["brand"]:
        parts.append(f"Brand: {row['brand']}")

    # Product type / category (your 'products' column)
    if row["products"]:
        parts.append(f"Product type: {row['products']}")

    # Color
    if row["baseColour"]:
        parts.append(f"Colour: {row['baseColour']}")

    # Price
    if pd.notna(row["price"]):
        parts.append(f"Price: {row['price']}")

    # Ratings (only if we have some)
    if row["ratingCount"] and row["ratingCount"] > 0:
        parts.append(
            f"Average rating: {row['avg_rating']} based on {row['ratingCount']} ratings"
        )

    # Main description
    if row["description"]:
        parts.append(f"Description: {row['description']}")

    # Extra attributes (size/fit/occasion/etc. often live here)
    if row["p_attributes"]:
        parts.append(f"Attributes: {row['p_attributes']}")

    # Join everything into a single text block
    return ". ".join(parts)

# Create the text column for embeddings
df["raw_text"] = df.apply(build_text, axis=1)
```

```python
#Save Cleaned data
processed_path = "data/processed/myntra_cleaned.csv"
os.makedirs(os.path.dirname(processed_path), exist_ok=True)
df.to_csv(processed_path, index=False)
processed_path
```
